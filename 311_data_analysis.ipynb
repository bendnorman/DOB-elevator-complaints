```
dob_boilr_df['Disposition Date'].dropna(inplace=True)
dob_boilr_df['Disposition Year'] = dob_boilr_df['Disposition Date'].str.split('/',2)
dob_boilr_df['Disposition Year'].dropna(inplace=True)
print(dob_boilr_df['Disposition Year'])

#dob_boilr_df['Disposition Year'] = dob_boilr_df['Disposition Year'][2]
```

```
time_diffs = pd.Series((pd.to_datetime(dob_boilr_df['Inspection Date']) - pd.to_datetime(dob_boilr_df['Date Entered'])).astype('timedelta64[D]'))
time_diffs = pd.DataFrame({'category': dob_boilr_df['Complaint Category'], 'time': time_diffs, 'Code': dob_boilr_df['Disposition Code']})
time_diffs.head()
```

```
print(len(time_diffs))
time_diffs = time_diffs[time_diffs['time'] >= 0]
print(len(time_diffs))
```

```
time_diffs = time_diffs[time_diffs['Code'].str.contains('A')]
print(len(time_diffs))
```
